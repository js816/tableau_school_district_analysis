```python
# Loading the dependencies
import pandas as pd
```

```python
# Loading the student file as a DataFrame
student_df = pd.read_csv('Resources/students_complete.csv')
student_df
```

```python
# Looking for null values
student_df.isnull().sum()
```

```python
# Looking at the summary statistics
student_df.describe()
```

```python
# Looking at the DataTypes
student_df.dtypes
```

```python
# Loading the school data as a DataFrame
school_df = pd.read_csv('Resources/schools_complete.csv')
school_df
```

```python
# Looking at the summary stattistics
school_df.describe()
```

```python
# Combining the data into DataFrame
complete_df = pd.merge(student_df, school_df, on=['school_name', 'school_name'])
complete_df
```

```python
# Creating budget per student column
school_df['budget_per_student'] = round(school_df['budget'] / school_df['size'], 0)
school_df
```

# Calculating some overall statistics

```python
student_count = complete_df['Student ID'].count()
student_count
```

```python
school_count = len(complete_df['school_name'].unique())
school_count
```

```python
total_budget = school_df['budget'].sum()
total_budget
```

```python
average_math_score = round(complete_df['math_score'].mean(),2)
average_math_score
```

```python
average_reading_score = round(complete_df['reading_score'].mean(),2)
average_reading_score
```

## Geting list of all students passing each test

```python
passing_math_df = complete_df[complete_df['math_score'] >= 70]
passing_math_df
```

```python
passing_reading_df = complete_df[complete_df['reading_score'] >= 70]
passing_reading_df
```

```python
## Calculating the number of students passing each test
```

```python
passing_math = passing_math_df['student_name'].count()
passing_math
```

```python
passing_reading = passing_reading_df['student_name'].count()
passing_reading
```

## Calculating the percent passing each test

```python
passing_math_percentage = round(passing_math / student_count * 100, 2)
passing_math_percentage
```

```python
passing_reading_percentage = round(passing_reading / student_count * 100, 2)
passing_reading_percentage
```

## Students who passed both tests

```python
# Getting DataFrame of students passing both
passing_both_df = complete_df[(complete_df['math_score'] >= 70) & (complete_df['reading_score'] >= 70)]
passing_both_df
```

```python
# Counting students passing both
passing_both = passing_both_df['Student ID'].count()
passing_both
```

```python
# Calculating percentage of students passing both
passing_both_percentage = round(passing_both / student_count * 100, 2)
passing_both_percentage
```

```python
# Creating DataFrame with district summary
district_summary_df = pd.DataFrame(
            [{'Total Schools': school_count,
              'Total Students': student_count,
              'Total Budget': total_budget,
              'Average Math Score': average_math_score,
              'Average Reading Score': average_reading_score,
              '% Passing Math': passing_math_percentage,
              '% Passing Reading': passing_reading_percentage,
              '% Passing Both': passing_both_percentage
             }])
district_summary_df
```

```python
# Formatting total students and total budget columns
district_summary_df['Total Students'] = district_summary_df['Total Students'].map('{:,}'.format)
district_summary_df['Total Budget'] = district_summary_df['Total Budget'].map('${:,}'.format)
district_summary_df
```
